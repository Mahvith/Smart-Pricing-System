```python
df = pd.read_csv(os.path.join(DATASET_FOLDER, 'train.csv'))
df.head()
```

```python
def extract_quantity(text):
    match = re.search(r'(\d+(\.\d+)?)\s*(oz|ounce|ml|g|kg|lb|lbs|l|litre|liter)', text.lower())
    if match:
        return float(match.group(1))
    return None


df['quantity_value'] = df['catalog_content'].apply(extract_quantity)
df.head()
```

```python
df.quantity_value.value_counts()
df.quantity_value.isnull().sum()
```

```python
# # Assuming df is your DataFrame with the 'quantity_value' column already created
# # (from the earlier extraction code)

# # Filter rows where quantity_value is null
missing_qty_df = df[df['quantity_value'].isnull()]

# # Print them
print("Number of products with missing quantity:", len(missing_qty_df))
print("\nSample rows with missing quantity_value:\n")
print(missing_qty_df[['sample_id', 'catalog_content']].iloc[1][1])
```

```python
# Extract numeric value after "Value:"
df["value"] = df["catalog_content"].str.extract(r"Value:\s*([\d\.]+)")

# Convert to float (optional)
df["value"] = df["value"].astype(float)
```

```python
df.head()
```

```python
missing_qty_df = df[df['value'].isnull()]

# # Print them
print("Number of products with missing quantity:", len(missing_qty_df))
# print("\nSample rows with missing value:\n")
print(missing_qty_df[['sample_id', 'catalog_content']].iloc[1][1])
missing_qty_df.head()
# missing_qty_df.isna().sum()
# missing_qty_df.isnull().sum()
```
